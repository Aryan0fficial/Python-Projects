```python
import numpy as np
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import plotly.express as px
import joblib
import pickle
```

```python
df = pd.read_excel(r'E:\Data Analytics\Python\Python Datas\Superstore.xlsx')
```

```python
df.shape
```

```python
df.isnull().mean()*100
```

```python
x = df[['Sales']]
y = df[['Profit']]

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.30, random_state=42)
```

```python
print('Total Data :',len(df))
print('x_train Data :',len(x_train))
print('x_test Data :',len(x_test))
```

```python
model = LinearRegression()
model
```

```python
model.fit(x_train,y_train)
```

```python
predicted = model.predict(x_test)

print('Predicted Data :',predicted[:10])
print('Actual Data :',y_test.values[:10])
```

```python
# convert Data frame
result = pd.DataFrame({'Sales':x_test.values.flatten(),
                      'Actual Profit':y_test.values.flatten(),
                      'Predicted Profit':predicted.flatten()})
```

```python
print('R2_train :',r2_score(y_train,model.predict(x_train)))
print('R2_test :',r2_score(y_test,predicted))
```
