```python
# Linear Regression = Future prediction karne ka tool

# Train Data =   Model ko sikhane ke liye
# Test Data =   Model ko check karne ke liye
# Prediction =  Train hone ke baad new values ka estimate
```

```python
x = df2[['total_bill','size','day','smoker']]
y = df2[['tip']]

x_train,x_test, y_train,y_test = train_test_split(x,y, test_size=0.20, random_state=42)
```

```python
print('Total Data :',len(df2))
print('Train Data :',len(x_train))
print('Test Data :',len(x_test))
```

```python
model = LinearRegression()
model
```

```python
model.fit(x_train,y_train)
```

```python
# Use the trained model to predict values on the test dataset and display the first 10 predicted and 
# actualvalues for comparison.

predicted = model.predict(x_test)

print('Predicted Value :',predicted[:10])
print('Actual Value :',y_test.values[:10])
```

```python
# Write Python code to display the R¬≤ score for both training and testing datasets to evaluate model performance

print('R2_train :',r2_score(y_train,model.predict(x_train)))
print('R2_test :',r2_score(y_test,predicted))
```
